
# Leccion 43 - Dropping and Filling NAs



```python
import pandas as pd
import numpy as np
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'] ,index_col=['country'], squeeze=True)
```

```python
#Revisamos los primeros registros y vemos que contiene valores NAN
alcohol.head()
```

```python
#Aplicamos el método dropna y revisamos nuevamente, los NAN ya no aparecen.
alcohol.dropna().head()
```

```python
#La ejecución de dropna no afecta la serie original, para modificarla podemos usar inplace. o reasignar a la misma variable
#alcohol = alcohol.dropna()
alcohol.dropna(inplace=True)
alcohol
```

```python
#También Podemos sustituir los NAN por algún valor más significativo con fillna()
alcohol.fillna('valor vacío', inplace=False)
```
