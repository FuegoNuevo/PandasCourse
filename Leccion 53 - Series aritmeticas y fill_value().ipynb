# Leccion 53 - Series aritmeticas y fill_value()

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# Podemos hacer operaciones aritmeticas a toda la Serie
# Por ejemplo agregar 2 bebidas a cada registro de la serie
alcohol + 2
```

```python
# Restar 10 y multiplicarlo por 2
(alcohol - 10) * 2
```

```python
# Es posible hacer operaciones entre series para eso creamos una nueva serie con un único registro
more_drinks = pd.Series({'Albania':6})
more_drinks
```

```python
alcohol + more_drinks
```

```python
# Intentamos nuevamente con dos paises nuevos
more_drinks = pd.Series({'Albania':6, 'Algeria':10})
alcohol + more_drinks
```

```python
# Intentamos con el método add
alcohol.add(more_drinks)
```

```python
# Dado que ocurre el mismo problema llenaremos el resto de valores con 0
alcohol.add(more_drinks, fill_value = 0)
```

```python
# Para el resto de ejemplos agregamos mas paises a la segunda Serie
more_drinks = pd.Series({'Albania':6, 'Algeria':10, 'Alberia' : 19, 'Afghanistan': 100, 'Yemen': 101})
```

```python
alcohol.subtract(more_drinks, fill_value = 0)
```

```python
alcohol.multiply(more_drinks, fill_value = 0)
```

```python
alcohol.divide(more_drinks, fill_value = 1)
```
