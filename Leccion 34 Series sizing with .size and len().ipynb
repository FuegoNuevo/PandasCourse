*   Atributos dan propiedades relacionadas al objeto
*   Metodos realizan una acción

```
import pandas as pd
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```
alcohol.size
```

```
alcohol.values
```

```
alcohol.index
```

```
# Cada uno de estos atributos tiene su propio atributo de tamaño size
alcohol.index.size == alcohol.size
```

```
# El atributo Shape nos permite ver el tamaño y el número de dimensiones de un DataFrame.
# En el caso de una Serie esta es unidimensional

alcohol.shape
```

```
# Dado que es una Tupla podemos obtener únicamente el primer elemento
alcohol.shape[0]
```

```
# Existe un método que nos permite obtener la longitud de la Serie
len(alcohol)
```
