# Leccion 50 - Ordenar con sort_index()

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# Es posible ordenar la seríe por los valores de su indice con el método sort_index()
alcohol.sort_index()
```

```python
alcohol.sort_index(ascending=False)
```

```python
alcohol.sort_index(ascending=False, na_position='first')
```

```python
alcohol.sort_index(ascending=False, na_position='first', inplace=True)
```
