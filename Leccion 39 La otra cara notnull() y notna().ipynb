```
import pandas as pd
import numpy as np
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```
# Al contrario de isnull(), el método notnull() refresa True si el valor es no nan
alcohol.notnull()
```

```
# Al igual que isnull pudemos usarlo para enmascarar la Serie o Dataframe.
# A modo de ilustración usamos loc que nos permite hacer selecciones con valores booleanos

# a.lcohol[alcohol.notnull()]
alcohol.loc[alcohol.notnull()]
```

```
# Buscamos cuántos registros no tienen valores nan

alcohol.notnull().sum()
```

```
# al sumar los resultados de notnull e isnull deberíamos obtener la totalidad de elementos en la Serie
alcohol.notnull().sum() + alcohol.isnull().sum() == alcohol.size
```
