```
import pandas as pd
import numpy as np
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

1. Aisla los valores no nulos de la serie alcohol  y asignalos a la variable wine_servings

```
wine_servings = alcohol[alcohol.notnull()]
wine_servings
```

2 . ¿Cuál es el total de vino consumido por pais en wine_servings?

```
wine_servings.sum()
```
