### Leccion 49 - nlargest() nsmallest()

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze = True)
```

```python
# Los métodos min y max son limitados pues sólo nos regresan el valor del elemento mayor.
alcohol.min()
```

```python
# Podemos usar max para obtenr los 10 numeros mayores de la siguiente forma:
alcohol.sort_values(ascending=False)[:10]
```

```python
# Un método más pythonico puede ser usando nlargest() y nsmallest()
alcohol.nlargest(10)
```

```python
alcohol.nsmallest(20)
```
