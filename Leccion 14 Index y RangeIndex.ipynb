## Leccion 14 Index y RangeIndex


```python
  import pandas as pd
```

```python
# como hemos visto Pandas asigna a cada elemento una etiqueta llamada indice
books_list = ['Fooled by Randomness', 'Sapiens', 'Lenin on the Train']
```

```python
list_s = pd.Series(books_list)
```

```python
# estas etiquetas inician con el numero cero y terminan en un entero n-1
list_s
```

```python
# Este indice puede ser declarado explicitamente con el par치metro index.
pd.Series(data=books_list, index=['funny', 'serious and amusing', 'kinda interesting'])
```

```python
# Tambien podemos usar unicamente los argumentos de forma posicional.
pd.Series(books_list, ['funny', 'serious and amusing', 'kinda interesting'])
```

```python
# A partir de la versi칩n 1.0 de Pandas, se instrodujo el tipo de dato string
pd.__version__
```

```python
# Asi que podemos declarar explicitamente el tipo de dato como string
# stringdtype
pd.Series(books_list, ['funny', 'serious and amusing', 'kinda interesting'], dtype='string')
```

```python
# Para generar el indice autom치tico, Pandas usa la funci칩n RangeIndex de la siguiente forma.
list_s.index
```

```python
type(list_s.index)
```

```python
# Podemos visualizar la salida de la funcion RangeIndex
# donde el valor de start es inclusivo pero el valor de stop es exclusivo
list(pd.RangeIndex(start=10, stop=25, step=5))
```

```python
# El indice creado por RangeIndex es inmutable, es decir que no puede ser modificado una vez creado por razones de performance.
```
