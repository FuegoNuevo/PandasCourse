# Leccion 57 - Iteración de series

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# Creamos una Serie de pruebas con 10 elementos
mini_alc = alcohol[:10]
```

```python
# Si queremos iterar sobre una Serie
for i in mini_alc:
  print(i)
```

```python
# Si queremos iterar sobre el indice:
for i in mini_alc.index:
  print(i)
```

```python
# Iterar sobre indices y valores
for i in mini_alc.index:
  print(i, mini_alc[i])
```

```python
# Iterar sobre los objetos
for i in mini_alc.items():
  print(i)
```
