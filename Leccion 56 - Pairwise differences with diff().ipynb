# Leccion 56 - Pairwise differences with diff()

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# diff() regresa una Serie con la diferencia del elemento menos su antecesor por default
alcohol.diff()
```

```python
# Podemos cambiar el antecesor con period en este ser√≠a elemento3 - elemento 1, elemento 4 - elemento 2, ...
alcohol.diff(periods = 2)
```

```python
# Podemos hacer un periodo negativo lo cual cambia el orden de los operandos, en este caso E1 - E2, E2- E3
alcohol.diff(periods = -1)
```
