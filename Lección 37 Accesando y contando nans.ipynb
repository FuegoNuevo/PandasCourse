```
import pandas as pd
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```
# Al comparar los valores de size y count pudimos ver que existen valores nan o null
alcohol.size == alcohol.count()
```

```
# Podemos ver que registros tienen valor null
alcohol.isnull()
```

```
# Usamos esos valores booleanos como máscara para regresar registros nan o null
alcohol[alcohol.isnull()]
```

```
# De esta seri podemos extraer el indice
alcohol[alcohol.isnull()].index
```

```
# Posteriormente lo convertimos a lista
list(alcohol[alcohol.isnull()].index)
```

```
# Para averiguar el número de registros con valor null o nan podemos usar
len(list(alcohol[alcohol.isnull()].index))
```

```
# Existen formas más siples de conseguir el mismo valor
alcohol.isnull().sum()
# sum regresa la suma de unos TRUE y ceros FALSE
```
