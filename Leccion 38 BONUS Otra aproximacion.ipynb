```
import pandas as pd
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```
# Otra forma para obtener los registros nan involucra numpy
import numpy as np
```

```
#ufunc -> Universal Function
np.isnan
```

```
ser = pd.Series(data=[True, False, None, 2], dtype=float)
ser
```

```
# usando la ufunc isnan tenemos
np.isnan(ser)
```

```
 # Usando isnan como una mascara
alcohol[np.isnan].size
```

```
# Asi mismo se puede usar la serie directamente
alcohol[alcohol.isna()].size
```
