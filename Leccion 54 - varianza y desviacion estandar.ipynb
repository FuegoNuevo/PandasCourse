# Leccion 54 - calculando la varianza y desviacion estandar

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# Podemos calcular la varianza con el método var()
alcohol.var()
```

```python
# Pero también podemos hacerlo usando operaciones aritméticas
# Usamos la fórmula de la varianza s^2 = sumatoria(X - media muestral)^2  / (n-1)
(alcohol.subtract(alcohol.mean())**2).sum() / (alcohol.count() - 1)
```

```python
# Para obtener la desviación estándar:
alcohol.std()
```

```python
# Calculandolo, sólo es necesario sacar la raiz cuadrada
((alcohol.subtract(alcohol.mean())**2).sum() / (alcohol.count() - 1))**(1/2)
```
