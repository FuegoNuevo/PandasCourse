El método count() se diferencía del atributo size ya que count() no considera los valores NAN o null

```
import pandas as pd
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```
# Si la cantidad de elementos reportada por size es diferente a count() quiere decir que existen valores nan
alcohol.size == alcohol.count()
```

```
# Verificando que contiene nan
alcohol.hasnans
```
