# Leccion 45 - El Método describe()

```python
import pandas as pd
import numpy as np
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'] ,index_col=['country'], squeeze=True)
```

```python
# Muchas de las medidas estdist|icas más comunes pueden nser obtenidas con el método describe()
alcohol.describe()
```

```python
# Podemos pasar algunos parámetros para personalizarlo
alcohol.describe(percentiles=(0.79, 0.19))
```

```python
alcohol.describe(include=float, exclude=object)
```
