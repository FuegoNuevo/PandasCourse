# Leccion 47 - idxmax() y idxmin()

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# Para obtener la etiqueta del valor máximo de una serie podemos usar una máscara booleana de la siguuientye forma:
alcohol[alcohol.values == alcohol.max()]
```

```python
# Para obtener el valor cualitativo podemos usar:
alcohol[alcohol.values == alcohol.max()].index[0]
```

```python
# O bien recurrimos a idxmax()
alcohol.idxmax()
```

```python
# Podemos usar idxmin() para obtener el indice del valor minimo
alcohol.idxmin()
```

```python
# Como recordamos, esta serie contiene 28 elementos de valor mínimo 1. 
# En este caso podemos ver el total de elementos de valor minimo con una máscara booleana
alcohol[alcohol.values == alcohol.min()]
```

```python
#Podemos interpretar una equivalencia como sigue:
alcohol[alcohol.idxmin()] == alcohol.min()
```
