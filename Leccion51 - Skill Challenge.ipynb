```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

# Leccion 51 Skill Challenge

### 1. Selecciona todos los paises de alcohol que tengan más de 50 servicios de vino y guardalos junto con sus valores correspondientes en una variable fifty_plus


```python
fifty_plus = alcohol[alcohol.values >= 50]
fifty_plus
```

### 2. De Fifty_plus, elige los paises con los 20 menores valores de  servicios de vino.

```python
smaller20 = fifty_plus.nsmallest(20)
smaller20
```

### 3. Cuál es la media, la mediana, y la desviación estandar de la muestra del paso 2



```python
smaller20.mean()
```

```python
smaller20.median()
```

```python
smaller20.std()
```

```python
smaller20.describe()
```
