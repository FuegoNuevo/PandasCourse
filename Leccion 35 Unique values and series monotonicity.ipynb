```
import pandas as pd
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```
# para revisar si la serie contiene valores unicos poder usar el atributo is_unique
alcohol.is_unique
```

```
# Si limitamos el resultado, el atributo is_unique también se ve afectado.
alcohol.head().is_unique
```

```
# La cantidad de valores unicos está dada por nunique(), este método no considera los NAN
alcohol.nunique()
```

```
# Para incluir los NAN se puede usar.
alcohol.nunique(dropna = False)
```

```
# La monotonicidad nos indica si los resultados de una serie son monotonamente ascendentes o monotonamente descendente.
pd.Series([1,2,3,4]).is_monotonic
```

```
pd.Series([1,2,13,4]).is_monotonic
```

```
# is monotonic es un alias de is_monotonic_increasing
pd.Series([1,2,3,4,5]).is_monotonic_increasing
```

```
# Tambien existe is_monotonic_decreasing
pd.Series(reversed([1,2,3,4,5])).is_monotonic_decreasing
```
