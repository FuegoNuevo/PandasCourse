# Lección 46 - mode() and value_counts()

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'] ,index_col=['country'], squeeze=True)
```

### Moda

```python
# Podemos calcular la moda con mode()
alcohol.mode()
#El resultado nos indica que el valor más común es 1.
```

```python
# Para saber cuánte vaces aparece el número 1 podemos usar una máscara booleana
alcohol[alcohol.values == 1].size
```

### Frecuencia con value_counts()



```python
# Podemos averiguar el número de ocurrencias para cada valor con value counts
alcohol.value_counts()
```

```python
# Para obtener la moda podemos usar value_counts() eligiendo el primer registro por la posición con iloc[]
alcohol.value_counts().iloc[0]
```
