# Leccion 55 - Cumulative Operations

```python
import pandas as pd
import numpy as np

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
```

```python
# Una suma acumulativa regresa una Serie que contiene el valor de lcada registro sumado a los valores que lo anteceden.
# Por default no considera los valores NAN.
alcohol.cumsum()
```

```python
# La suma de cualquier valor con un NA es igual a NA
```

```python
# Multiplicación acumulada
alcohol.cumprod()
```

```python
# cummin devuelve el valor más pequeño hasta el momento.
alcohol.cummin()
```

```python
# cummax devuelve el valor más pequeño hasta el momento.
alcohol.cummax()
```
