#Leccion 48 - Ordenando con sort_values()

```python
import pandas as pd
import numpy as np
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols = ['country', 'wine_servings'], index_col=['country'], squeeze = True)
```

```python
#Podemos ordenar los valores de una serie con el método sort_values()
alcohol.sort_values()
```

```python
# Podemos invertir el orden ascendente por descendente
alcohol.sort_values(ascending = False)
```

```python
# Podemos cambiar la posición de los nan
alcohol.sort_values(ascending = False, na_position='first')
```

```python
# podemos especificar el tipo de alrgoritmo usado pada ordenar, el default es quicksort
alcohol.sort_values(ascending = False, kind = 'quicksort') #podemos usar mergesort y heapsort
```

```python
# también podemos asignar la Serie ordenada a la misma variable en este caso alcohol
alcohol.sort_values(ascending = False, inplace=True)
alcohol
```
