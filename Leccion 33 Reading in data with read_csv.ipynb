```
import pandas as pd
```

Leyendo datos con read_csv()

CSV significa Comma Separated Values

```
# El archivo csv para este ejercicio se encuentra en la siguiente ruta.
# https://andybek.com/pandas-drinks

pd.read_csv('https://andybek.com/pandas-drinks').head()
```

```
# Limitamos el número de columnas
pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'])
```

```
# Usando la columna country como indice 
# Y asignando el resultado a la variabe alcohol
alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'])
```

```
# Como vemos el tipo de dato que regresa es un DataFrame. 
type(alcohol)
```

```
# Existe un parámetro de la función read que permite regresar una Serie 
# siempre que el Dataframe tenga una única columna.

alcohol = pd.read_csv('https://andybek.com/pandas-drinks', usecols=['country', 'wine_servings'], index_col=['country'], squeeze=True)
type(alcohol)
```
